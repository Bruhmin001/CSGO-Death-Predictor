```python
def vif_checker(df,y,cols2drop : list):

    cols2drop.append(y)
    X = add_constant(df.drop(cols2drop, axis=1))

    vif_df = pd.DataFrame(
        {
            'features' : X.columns,
            'VIF' : [variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
        })
    return vif_df
```
